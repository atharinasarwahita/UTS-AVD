#### **4.2.8 Cek Outliers**

```python
results = []

cols_numeric = df.select_dtypes(include=['float64', 'int64']).columns

for col in cols_numeric:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

display(results_df)
```

Pada pengecekan outliers, tidak terdapat outliers pada kolom numerik. Sehingga tidak tercatat adanya data dengan nilai yang ekstrem atau kesalahan pengisian data.

### **4.4 Exploratory Data Analysis (EDA)**

#### **4.4.1 Perbandingan (Comparison)**

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle('Rata-rata Final Score berdasarkan Faktor Kategorikal', fontsize=14)

# 1. Gender
df.groupby('gender')['final_score'].mean().plot(kind='bar', ax=axes[0], color=['#FF9999','#66B2FF'], edgecolor='black')
axes[0].set_title('Final Score by Gender')
axes[0].set_ylabel('Rata-rata Final Score')
axes[0].tick_params(axis='x', rotation=0)

# 2. Internet Access
df.groupby('internet_access')['final_score'].mean().plot(kind='bar', ax=axes[1], color=['#FF9999','#66FF99'], edgecolor='black')
axes[1].set_title('Final Score by Internet Access')
axes[1].tick_params(axis='x', rotation=0)

# 3. Extracurricular
df.groupby('extracurricular')['final_score'].mean().plot(kind='bar', ax=axes[2], color=['#FFD700','#DA70D6'], edgecolor='black')
axes[2].set_title('Final Score by Extracurricular')
axes[2].tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.show()
```

Ketiga grafik batang menampilkan rata-rata nilai akhir berdasarkan gender, akses internet, dan keikutsertaan ekstrakurikuler. Dari ketiga faktor tersebut, tidak ada satu pun yang menunjukkan perbedaan berarti antar kategorinya,seluruh batang berada di kisaran nilai yang hampir sama, sekitar 55–57.

*   Gender: Siswa laki-laki dan perempuan memiliki rata-rata nilai akhir yang hampir identik. Artinya, gender tidak menjadi faktor pembeda performa akademik dalam dataset ini.


*   Akses Internet: Siswa dengan dan tanpa akses internet menunjukkan rata-rata nilai yang setara. Ketersediaan internet tidak terbukti memberikan keunggulan akademik yang signifikan.

*   Ekstrakurikuler: Siswa yang mengikuti ekstrakurikuler maupun yang tidak memiliki rata-rata nilai yang relatif sama, mengindikasikan bahwa keikutsertaan ekstrakurikuler tidak mengganggu maupun meningkatkan performa secara berarti.

Ketiga faktor kategorikal ini, yaitu gender, akses internet, dan ekstrakurikuler bukan merupakan faktor penentu performa akademik siswa. Fokus analisis sebaiknya diarahkan ke variabel lain yang lebih berpengaruh, seperti jam belajar per minggu.

#### **4.4.2 Komposisi (Composition)**

```python
passed = df['passed'].value_counts()
passed_resp = passed.head(4)

plt.figure(figsize=(8, 6))
passed_resp.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Proporsi Siswa Lulus dan Tidak Lulus')
plt.ylabel('')
plt.axis('equal')
plt.show()
```

Pie chart menunjukkan bahwa 64,6% siswa dinyatakan lulus dan 35,4% tidak lulus.

Meskipun mayoritas siswa berhasil lulus, proporsi siswa yang tidak lulus sebesar 35,4% terbilang cukup besar dan tidak bisa diabaikan. Hampir 1 dari 3 siswa gagal mencapai kelulusan, yang mengindikasikan masih adanya kesenjangan performa yang perlu ditangani. Temuan ini memperkuat urgensi untuk mengidentifikasi faktor-faktor yang membedakan siswa lulus dan tidak lulus yang dalam analisis ini mengerucut pada intensitas belajar sebagai variabel paling dominan.

#### **4.4.3 Distribusi (Distribution)**

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['final_score'], bins=20, kde=True)
plt.title('Distribusi Final Score')
plt.xlabel('Jumlah Kafein (mg)')
plt.ylabel('Frekuensi')
plt.show()
```

Distribusi final score mendekati distribusi normal (simetris), dengan puncak di sekitar nilai 55–60. Tidak terdapat kemiringan (*skewness*) ekstrem, yang mengindikasikan bahwa nilai siswa tersebar cukup merata tanpa ada pengelompokan yang mencolok di sisi rendah maupun tinggi.

#### **4.4.4 Hubungan (Relationship)**

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data=df[['age','study_hours_per_week','attendance_rate','previous_score','final_score']].corr(),
            annot=True,
            cmap='coolwarm',
            fmt='.2f')
plt.title('Heatmap Korelasi')
plt.show()
```

Kolom study_hours_per_week memiliki korelasi 0,80 dengan final_score, ini merupakan angka korelasi tertinggi dan paling dominan yang berarti sekitar 64% variasi nilai akhir dapat dijelaskan oleh jam belajar per minggu saja. Sebaliknya, usia (*age*) hampir tidak memiliki hubungan dengan nilai akhir.
